# Conditional Statements in Python
Definition

Conditional Statements are programming statements that allow a program to make decisions based on whether a condition is True or False. They help the program execute different blocks of code depending on the result of a condition.

## What is a Condition?

A condition is an expression that always returns either True or False.

Example

```python
10 > 5
```

## What is a Boolean Expression?

A Boolean Expression is an expression that evaluates to either True or False.

Example

```python
age = 18
age >=  18
```

## Types of Conditional Statements
if Statement

if-else Statement

if-elif-else Statement

Nested if Statement

## if Statement
Definition

The if statement executes a block of code only when the given condition is True. If the condition is False, the block is skipped.

### Syntax
if condition:

    statement

```python
age =  20
if age >= 18:
    print("Eligible to vote")
    
```
