# Conditional Statements in Python
Definition

Conditional Statements are programming statements that allow a program to make decisions based on whether a condition is True or False. They help the program execute different blocks of code depending on the result of a condition.

## What is a Condition?

A condition is an expression that always returns either True or False.

Example

```python
10 > 5
```

## What is a Boolean Expression?

A Boolean Expression is an expression that evaluates to either True or False.

Example

```python
age = 18
age >=  18
```

## Types of Conditional Statements
if Statement

if-else Statement

if-elif-else Statement

Nested if Statement

## if Statement
Definition

The if statement executes a block of code only when the given condition is True. If the condition is False, the block is skipped.

### Syntax

    statement

```python
age =  20
if age >= 18:
    print("Eligible to vote")
    
```

# if-else Statement
Definition

The if-else statement is used to perform two-way decision making. If the condition is True, the if block is executed. Otherwise, the else block is executed. Only one block executes at a time.

Syntax

if condition:

    # Code if condition is True
    
else:

    # Code if condition is False

## if-elif-else Statement
Definition

The if-elif-else statement is used when there are multiple conditions to check. Python evaluates the conditions from top to bottom. When a condition is True, its corresponding block is executed, and the remaining conditions are skipped. If none of the conditions are True, the else block is executed.

Syntax

if condition1:

    # Code Block 1

elif condition2:

    # Code Block 2

elif condition3:

    # Code Block 3

else:

    # Default Code Block

```python
#1: Grade Calculator (A, B, C, D, Fail)
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
elif marks >= 35:
    print("Grade D")
else:
    print("Fail")
```

##  Nested if Statement
Definition

A Nested if statement is an if statement inside another if statement. The inner if statement is executed only if the outer if condition is True. It is used when one condition depends on another condition.

Syntax
    
if condition1:

    # Outer if block

    if condition2:
        
        # Inner if block
        
How Nested if Works
    
Python checks the outer if condition.
    
If the outer condition is True, it checks the inner if condition.
    
If the inner condition is also True, the inner block is executed.
    
If the outer condition is False, the inner if is skipped.

```python
#1: Student Exam Eligibility
attendance = int(input("Enter attendance percentage: "))

if attendance >= 75:
    fees = input("Are fees paid? (yes/no): ")

    if fees == "yes":
        print("Eligible for Exam")
    else:
        print("Pay Fees First")
else:
    print("Attendance is Less Than 75%")
```

```python
#2: Login System
username = input("Enter Username: ")

if username == "admin":
    password = input("Enter Password: ")

    if password == "1234":
        print("Login Successful")
    else:
        print("Incorrect Password")
else:
    print("Invalid Username")
    

```

```python
#3: ATM System
card = input("Is Card Inserted? (yes/no): ")

if card == "yes":
    pin = int(input("Enter PIN: "))

    if pin == 1234:
        print("Access Granted")
    else:
        print("Incorrect PIN")
else:
    print("Insert Your Card")
```

```python
#4: College Admission
marks = int(input("Enter Marks: "))

if marks >= 60:
    documents = input("Documents Verified? (yes/no): ")

    if documents == "yes":
        print("Admission Confirmed")
    else:
        print("Verify Documents")
else:
    print("Not Eligible for Admission")
```

```python
#5: Online Shopping
stock = int(input("Enter Product Quantity: "))

if stock > 0:
    payment = input("Payment Successful? (yes/no): ")

    if payment == "yes":
        print("Order Confirmed")
    else:
        print("Payment Failed")
else:
    print("Product Out of Stock")
```

```python
#6: Hospital Appointment
registered = input("Is Patient Registered? (yes/no): ")

if registered == "yes":
    doctor = input("Is Doctor Available? (yes/no): ")

    if doctor == "yes":
        print("Appointment Confirmed")
    else:
        print("Doctor Not Available")
else:
    print("Please Register First")
```

```python
#7: Nested if vs Logical Operator (and)
attendance = 80
fees_paid = True

if attendance >= 75:
    if fees_paid:
        print("Eligible for Exam")
    else:
        print("Fees Not Paid")
else:
    print("Attendance Short")

```
