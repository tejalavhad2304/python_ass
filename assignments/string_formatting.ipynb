#  String Formatting
Definition

String Formatting is the process of inserting values into a string in a readable and meaningful way.

## 1. Concatenation
Definition

Joining two or more strings using the + operator.

Syntax

string1 + string2

```python
name = "Tejal"
print("Hello "+name)
```

```python
#number example
age = 22
print("age :"+str(age))
```

## 2. Comma Separation
Definition

Using commas inside the print() function.

Example

```python
age=22
print("age :",age)
```

## 3. % Formatting
Definition

Old method of formatting strings using placeholders.

Example

```python
name="Tejal"
print("hello %s"%name)
```

```python
#multiple variales
name="tejal"
age=22
print("name: %s age :%d"%(name,age))
```

## 4. format()
Definition

The format() method inserts values into placeholders {}.

Example

```python
name = "tejal"
print("hello {}".format(name))
```

```python
name="Tejal"
age=22

print("Name : {} Age : {}".format(name,age))
```

## 5. f-Strings
Definition

f-Strings are the modern and recommended way to format strings.

Syntax

f"{variable}"

```python
name = "Tejal"
age  =  22
print(f"My name is {name} and  I  am {age}  years old")
```
