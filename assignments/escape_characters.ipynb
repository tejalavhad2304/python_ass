# Escape Characters
Definition

Escape characters are special character combinations that begin with a backslash (\). They allow special characters to be included inside strings.

## \n
Definition

Moves the cursor to the next line.

Example

```python
print("Hello\nWorld")
```

## \t
Definition

Adds horizontal space (tab).

Example

```python
print("Name\t\tage")
```

```python
print("Name\tage")
```

## \
Definition

Prints a backslash.

Example


```python
print("C:\\Users\\TEJAL\\OneDrive\\Desktop\\py")
```

## "
Definition

Prints double quotation marks inside a string.

Example

```python
print("She said \"Hello\"")
```

## '
Definition

Prints a single quotation mark inside a string.

Example

```python
print("It\'s me")
```
