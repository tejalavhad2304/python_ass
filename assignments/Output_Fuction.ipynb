## Output Function
Definition

Output is the information that a program sends to the user. In Python, the print() function is used to display output on the screen.

Syntax<br>
print(object1, object2, ..., sep=' ', end='\n')

## Parameters
object → Value(s) to display.

sep → Separator between multiple values (default = space).

end → Character printed at the end (default = new line).

```python
print("Hello World")
```

```python
name="Tejal"
print("Welcome",name)
```

## sep Parameter
Definition

The sep parameter specifies the separator between multiple values.

Syntax

print(value1,value2,sep="-")

```python
print("Python","Java","C++",sep="-")
```

```python
print(10,20,30,sep="*")
```

##  end Parameter
Definition

The end parameter specifies what should be printed at the end of the output.

Default value is '\n' (new line).

Example

```python
print("Hello",end="   ")
print("World")
```
