# Truthy & Falsy Values
Definition

In Python, every value is considered either Truthy or Falsy when used in a condition.

Truthy Value → Evaluates to True

Falsy Value → Evaluates to False

Truthy and Falsy values are commonly used in conditional statements to check whether a value exists or not.

## Common Truthy Values
Non-zero numbers (1, 5, -10)

Non-empty string ("Python")

Non-empty list ([1,2])

Non-empty tuple ((10,20))

Non-empty dictionary ({"name":"Tejal"})

True

## Common Falsy Values
0

0.0

""

[]

()

{}

None

False

### Syntax
if value:

    print("Truthy")
    
else:

    print("Falsy")

```python
name = "Tejal"

if name:
    print("Data Found")
else:
    print("No Data")
```

```python
marks = []

if marks:
    print("marks available")
else:
    print("No marks found")
```

## Match-Case Statement
Definition

The match-case statement is used to compare one value against multiple cases. It is similar to the switch statement in other programming languages.

Introduced in Python 3.10.

Syntax

match variable:

    case value1:
    
        statement

    case value2:
    
        statement

    case _:
    
        default statement


```python
day = "monday"

match day:
    case "monday":
        print("Working Day")
    case "sunday":
        print("Holiday")
    case _:
        print("Normal Day")
        
    
```

### Assignment 7 : Banking Deposit System
Question

Create a banking deposit program.

Input

Account Active (True/False)

KYC Completed (True/False)

Deposit Amount

Rules

If the account is inactive → Reject Deposit

If KYC is incomplete → Complete KYC

Deposit Amount ≤ 0 → Invalid Deposit Amount

Deposit Amount > ₹2,00,000 → Manager Approval Required

Deposit Amount > ₹50,000 → PAN Verification Required

Otherwise → Deposit Successful

```python
account_active = input("Is  account is  active? (True/False):")== "True"
kyc_completed = input("Is  KYC Completed? (True/False):")== "True"
amount = float(input("Enter Deposit Amount:"))

if not account_active:
    print("Deposit Rejected. Account is Inactive.")

elif not kyc_completed:
    print("Please Complete KYC.")

elif amount <= 0:
    print("Invalid Deposit Amount.")

elif amount > 200000:
    print("Manager Approval Required.")

elif amount > 50000:
    print("PAN Verification Required.")

else:
    print("Deposit Successful.")

```

### Assignment 11 : Mini Banking Application
Question

Create a banking application using match-case, conditional statements, and Truthy/Falsy values.

Menu

Deposit

Withdraw

Balance Enquiry

Exit

```python
balance = 10000

print("===== BANK MENU =====")
print("1. Deposit")
print("2. Withdraw")
print("3. Balance Enquiry")
print("4. Exit")

choice = int(input("Enter Your Choice: "))

match choice:

    case 1:
        amount = float(input("Enter Deposit Amount: ₹"))

        if amount > 0:
            balance += amount
            print("Deposit Successful.")
            print("Updated Balance = ₹", balance)
        else:
            print("Invalid Amount")

    case 2:
        amount = float(input("Enter Withdraw Amount: ₹"))

        if amount <= 0:
            print("Invalid Amount")

        elif amount > balance:
            print("Insufficient Balance")

        else:
            balance -= amount
            print("Withdrawal Successful.")
            print("Remaining Balance = ₹", balance)

    case 3:
        if balance:
            print("Current Balance = ₹", balance)
        else:
            print("No Balance Available")

    case 4:
        print("Thank You For Banking With Us.")

    case _:
        print("Invalid Choice")
```
