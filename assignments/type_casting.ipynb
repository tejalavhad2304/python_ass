# Type Casting
Definition

Type Casting is the process of converting one data type into another data type.

1.Why Type Casting is Required?

-The input() function always stores data as a string.

-Therefore, mathematical operations cannot be performed directly.

```python
age=input("Enter Age : ")
print(type(age))
```

```python
age=int(input("Enter Age : "))
print(age+1)
```

## int()
Definition

Converts a compatible value into an integer.

Example

```python
print(int("100"))
```

```python
print(int(20.5))
```

## Float()
Definition

Converts a compatible value into a floating-point number.

Example

```python
print(float(25))
```

```python
print(float(20.1))
```

## str()
Definition

Converts any data type into a string.

Example

```python
print(str(100))
```

```python
name  =  "Tejal"
age   =  22
print(name+"  "+str(age))
```

## bool()
Definition

Converts a value into either True or False.

Examples

```python
print(bool(0))
```

```python
print(bool(1))
```

```python
print(bool("  "))
```

```python
print(bool("python"))
```
