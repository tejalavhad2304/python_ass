<h3>5. Membership Operators</h3>
Membership operators are used to check whether a value exists inside a sequence or collection.

Collections include:

List

Tuple

Set

Dictionary

String

<h5>Type</h5>


in	    ->       Returns True if value exists

not in	  ->     Returns True if value does not exist

```python
numbers = [10,20,30,40]
print(20 in numbers)
print(50 in numbers)
```

```python
numbers = [10,20,30]
print(50 not in numbers)
print(20 not in numbers)
```

```python
text = "Python"
print("P" in text)
print("p" in text)
print("java" in text)
```

```python
t = (1,2,3,4)
print(1 in t)
print(5 in t)
print(5 not in t)
```

```python
student = {
    "name":"Rahul",
    "age":20
}

print("name" in student)
print("Rahul" in student)
```

```python
print("Rahul" in student.values())
```

<h3>6. Identity Operators</h3>
Identity operators compare whether two variables refer to the same object in memory, not just whether their values are equal.

<h5>Types</h5>
<pre>
Operator	Description
is	     ->   Returns True if both variables refer to the same object
is not	 ->   Returns True if they refer to different objects</pre>

<h5>- Difference Between == and is</h5>
<pre>
==  ->   compare values

is  ->   Compares memory location (object identity)</pre>

```python
a = [1,2,3]
b = a

print(a is b)
```

```python
a = [1,2,3]
b = [1,2,3]
print(a is b)
print(a == b)
```

```python
x = 50
y = 50

print(x is y)
```

```python
x = [10]
y = [10]

print(x is not y)
```

<h5>5. Left Shift (<<)</h5>

Moves bits to the left.

Each left shift by 1 is equivalent to multiplying by 2.

```python
print(5 << 2)
```

Binary

5 = 0101

0101 << 2

010100

<h5>6. Right Shift (>>)</h5>

Moves bits to the right.

Each right shift by 1 is approximately equivalent to floor division by 2 for positive integers.

```python
print(20 >> 2)
```

Binary

20 = 10100

10100 >> 2

00101

Output

5

Calculation

20 // 2² = 5
