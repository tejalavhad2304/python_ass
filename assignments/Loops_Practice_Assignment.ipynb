## Topic 1 : range() Function
Definition

The range() function is a built-in Python function used to generate a sequence of numbers. It is mainly used with loops when we need to repeat a task a fixed number of times.

### Why do we use range()?
- To generate numbers automatically.
- To control loop iterations.
- To print numbers in ascending or descending order.
- To create tables and patterns.
#### Syntax
range(stop)

range(start, stop)

range(start, stop, step)
##### Example
for i in range(1,6):

    print(i)

Output

1
2
3
4
5

```python
#Print numbers from 1 to 20.
for i in range(1,21):
    print(i,end="  ")
```

```python
#Print numbers from 20 to 1.
for i  in range(20,1,-1):
    print(i,end=" ")
```

```python
#Print all even numbers between 1 and 50.
for i in range(0,51,2):
    #if i % 2 == 0:
    print(i,end=" ")
```

```python
#Print all odd numbers between 1 and 50.
for i in range(1,50,2):
    print(i,end=" ")
    
```

```python
#Print the multiplication table of 7 using range().
for i in range(7,71,7):
    print(i,end = " ")
```

```python
#Print the multiplication table of 7 using range().
num = 7
for i in range(1,11):
    print(f"{num} x {i} =",num*i)
```

```python
for i in range(1,11):
    print(f"7x{i}={7*i}")
```

```python
#Print numbers from 100 to 10 with a gap of 10.
for i in range(100,9,-10):
    print(i)
```
