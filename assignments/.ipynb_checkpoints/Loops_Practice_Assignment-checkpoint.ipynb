## Topic 1 : range() Function
Definition

The range() function is a built-in Python function used to generate a sequence of numbers. It is mainly used with loops when we need to repeat a task a fixed number of times.

### Why do we use range()?
- To generate numbers automatically.
- To control loop iterations.
- To print numbers in ascending or descending order.
- To create tables and patterns.
#### Syntax
range(stop)

range(start, stop)

range(start, stop, step)
##### Example
for i in range(1,6):

    print(i)

Output

1
2
3
4
5

```python
#Print numbers from 1 to 20.
for i in range(1,21):
    print(i,end="  ")
```

```python
#Print numbers from 20 to 1.
for i  in range(20,1,-1):
    print(i,end=" ")
```

```python
#Print all even numbers between 1 and 50.
for i in range(0,51,2):
    #if i % 2 == 0:
    print(i,end=" ")
```

```python
#Print all odd numbers between 1 and 50.
for i in range(1,50,2):
    print(i,end=" ")
    
```

```python
#Print the multiplication table of 7 using range().
for i in range(7,71,7):
    print(i,end = " ")
```

```python
#Print the multiplication table of 7 using range().
num = 7
for i in range(1,11):
    print(f"{num} x {i} =",num*i)
```

```python
for i in range(1,11):
    print(f"7x{i}={7*i}")
```

```python
#Print numbers from 100 to 10 with a gap of 10.
for i in range(100,9,-10):
    print(i)
```

```python
#Print all multiples of 5 between 1 and 100. 
for i in range(5,100,5):
    print(i,end=" ")
```

## Assignment 2: for Loop

## Definition

A for loop is a control flow statement in Python used to execute a block of code repeatedly for each item in a sequence, such as a list, string, tuple, or range. It is mainly used when the number of iterations is known.

### Why do we use a for loop?
To repeat a block of code multiple times.

To iterate through a sequence (list, tuple, string, range, etc.).

To reduce repetitive code.

To perform calculations, printing, searching, and pattern generation.

### Syntax

for variable in sequence:

    # Code to execute
    
Using range()

for i in range(start, stop, step):

    print(i)
    
### Flow of a For Loop

The loop starts.

It takes the first value from the sequence

Executes the loop body.

Moves to the next value.

Repeats until all values are processed.

The loop ends.

```python
#Print your name 10 times.
for i in range(10):
    print("Tejal Avhad")
```

```python
#Print numbers from 1 to 100.
for i in range(1,101):
    print(i,end=" ")
```

```python
#Print the square of numbers from 1 to 10. 
for i  in  range(1,11):
    print(i,"=",i**2)
```

```python
#Print the cube of numbers from 1 to 10. 
for i  in  range(1,11):
    print(i,"=",i**3)
```

```python
#Find the sum of numbers from 1 to 100. 
sum=0
for i in range(1,101):
    sum=sum+i
    print(sum,end=" ")
```
