# Truthy & Falsy Values
Definition

In Python, every value is considered either Truthy or Falsy when used in a condition.

Truthy Value → Evaluates to True

Falsy Value → Evaluates to False

Truthy and Falsy values are commonly used in conditional statements to check whether a value exists or not.

## Common Truthy Values
Non-zero numbers (1, 5, -10)

Non-empty string ("Python")

Non-empty list ([1,2])

Non-empty tuple ((10,20))

Non-empty dictionary ({"name":"Tejal"})

True

## Common Falsy Values
0

0.0

""

[]

()

{}

None

False

### Syntax
if value:

    print("Truthy")
    
else:

    print("Falsy")

```python
name = "Tejal"

if name:
    print("Data Found")
else:
    print("No Data")
```

```python
marks = []

if marks:
    print("marks available")
else:
    print("No marks found")
```

## Match-Case Statement
Definition

The match-case statement is used to compare one value against multiple cases. It is similar to the switch statement in other programming languages.

Introduced in Python 3.10.

Syntax

match variable:

    case value1:
    
        statement

    case value2:
    
        statement

    case _:
    
        default statement


```python
day = "monday"

match day:
    case "monday":
        print("Working Day")
    case "sunday":
        print("Holiday")
    case _:
        print("Normal Day")
        
    
```

### Assignment 2 : Traffic Signal

Question

Create a Traffic Signal Program.

Input

Signal Color

Cases

Red → Stop

Yellow → Get Ready

Green → Go

Any Other → Invalid Signal

```python
signal = input("Enter signal coclor")

match signal:
    case "Red":
        print("Stop")
    case "Yellow":
        print("Get Ready")
    case "Green":
        print("Go")
    case _:
        print("Invalid Signal")
```

## Assignment 3 : Calculator
Question

Menu

1 Addition

2 Subtraction

3 Multiplication

4 Division

```python
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")

choice =int(input("Enter your choice:"))


num1 =float(input("Enter first number:"))
num2 =float(input("Enter second number:"))

match choice:
    case 1:
        print("Addition = ",num1 + num2)
    case 2:
        print("Substraction = ",num1 - num2)
    case 3:
        print("Multiplication = ",num1  * num2)
    case 4:
        if num2 != 0:
            print("Division =", num1 / num2)
        else:
            print("Cannot Divide by Zero")
    case _:
        print("Invalid Choice")
                  
             
                   
```
