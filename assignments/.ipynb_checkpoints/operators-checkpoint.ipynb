<h1>Operators</h1>
In Python, operators are special symbols used to perform operations on variables and values.

<h3>Type of Operators</h3>

1.Arithmetic Operators

2.Comparison (Relational) Operators

3.Logical Operators

4.Assignment  operators

<h2>1.Arithmetic Operators</h2>

Used for mathematical calculations.


1.+	Addition	

2.-	Subtraction	

3.*	Multiplication

4./	Division	

5.//	Floor Division

6.%	Modulus (Remainder)	

7.**	Exponent



```python
a = 10
b = 3

print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)   
print(a // b)  
print(a % b)   
print(a ** b)  
```

```python
x = 3.5
y = 7.2

print(x + y)   
print(x - y)   
print(x * y)   
print(x / y)   
print(x // y)  
print(x % y)   
print(x ** y)
```

```python
p = 2
q = -10

print(p+ q)   
print(p - q)   
print(p * q)   
print(p / q)   
print(p // q)  
print(p % q)   
print(p ** q)
```

<h2>2.Comparison operator</h2>

Comparison operator compare two values

They always return a Boolean value (True or False)

They are also called Rational Operators

They are used for decision making

Compare numbers ,strings,and many other obj


1.Equal to(==)

2.Not equal to(!=)

3.Greater than (>) 

4.Less than (<)

5.Greater than or equal to(>=)

6.Less than or equal to(<=)

```python
x = 5
y = 10
z = 5
print(x == y)
print(x == z)   
print(x != y)   
print(x < y)
print(x > y)   
print(x >= y)
print(x <= y)
```

```python
x = 5.12
y = -2.23

print(x == y)   
print(x != y)   
print(x < y)
print(x > y)   
print(x >= y)
print(x <= y)
```

```python
x = -3
y = -1

print(x == y)   
print(x != y)   
print(x < y)
print(x > y)   
print(x >= y)
print(x <= y)
```

<h2> 3.Logical Operators</h2>

Logical operators combine multiple conditions

They work with both boolean values(True and False)

python provide three logical operators:


1.and =	True if both conditions are true

2.or  =	True if at least one condition is true

3.not =	Reverses the result




```python
a = 5
print(a > 2 and a < 10)  
print(a > 10 or a < 10)  
print(not(a > 2))        
```

```python
a = 5
b =-2
print(b > 2 and a < 10)  
print(a > 10 or a < 10)  
print(not(a > 2))        
```
