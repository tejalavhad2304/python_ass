<h1>DATATYPE</h1>

A Data Type in Python defines the type of value a variable can store. Every value in Python belongs to a specific data type, and the data type determines what operations can be performed on that value.

<h3>Definition</h3>

<h5>A data type is a classification that specifies:</h5>

1.What kind of value is stored.

2.How the value is represented in memory.

3.What operations can be performed on it.

<h4>example</h4>

```python
x = 10
print(type(x))
```
