<h1>DATATYPE</h1>

A Data Type in Python defines the type of value a variable can store. Every value in Python belongs to a specific data type, and the data type determines what operations can be performed on that value.

<h3>Definition</h3>

<h5>A data type is a classification that specifies:</h5>

1.What kind of value is stored.

2.How the value is represented in memory.

3.What operations can be performed on it.

<h4>example</h4>

```python
x = 10
print(type(x))
```

<h5>Categories of Data Types in Python</h5>

Numeric

<h4>1. Numeric Data Types</h4>

Numeric types store numbers.

<h5>A) Integer (int)</h5>
Definition

An integer is a whole number without a decimal point.

Examples

```python
a = 10
b = -20
c = 0
```

```python
a = 20
b = 5

print(a + b)
print(a - b)
print(a * b)
print(a // b)
```

### Real-life Example
Age

Number of students

Number of books

## B) Float (float)
Definition

A float stores decimal numbers.

Example

```python
price = 99.99
temperature = 36.5

print(price)
print(temperature)
```

### Real-life Example
Height

Weight

Temperature

Product price

##  C) Complex (complex)
Definition

Complex numbers contain a real part and an imaginary part.

Format

a + bj

Example

```python
x = 3 + 5j

print(x)
print(type(x))
```

### Real-life Example

Used in:

Electrical engineering

Signal processing

Scientific calculations

## 2. Boolean Data Type (bool)
Definition

Stores only two values:

True

False

Example

```python
a = True
b = False

print(a)
print(b)
```

```python
#Example using comparison
x = 20

print(x > 10)
print(x < 10)
```

###  Real-life Example

Login successful

Door open/closed

Light ON/OFF

## 3. String (str)
Definition

A string is a sequence of characters enclosed in:

Single quotes

Double quotes

Triple quotes

Example

```python
name = "Tejal"

print(name)
print(type(name))
```

```python
name = "Python"

print(len(name))
print(name.upper())
print(name.lower())
print(name[0])
```

###  Real-life Example
Name

Address

Email

City

##  4. List (list)
Definition

A list is an ordered, mutable (changeable) collection that allows duplicate elements.

Example

```python
fruits = ["Apple", "Banana", "Mango"]

print(fruits)
```

```python
#Modifying a List
fruits[1] = "Orange"

print(fruits)
```

## Features
Ordered

Mutable

Allows duplicates

Indexed

### Real-life Example
Shopping list

Student names

Employee records

## 5. Tuple (tuple)
Definition

A tuple is an ordered, immutable (unchangeable) collection.

Example

```python
numbers = (10, 20, 30)

print(numbers)
```

### Features
Ordered

Immutable

Allows duplicates

Indexed

### Real-life Example

Latitude and longitude

RGB color values

Fixed configuration settings

## 6. Set (set)
Definition

A set is an unordered collection of unique elements.

Example

```python
#Duplicate values are automatically removed.
numbers = {10, 20, 30, 20}

print(numbers)
```

###  Features
Unordered

Mutable

No duplicates

Not indexed

###  Real-life Example
Unique student IDs

Unique email addresses

Unique tags 

## 7. Dictionary (dict)
Definition

Stores data in key-value pairs.

Example

```python
student = {
    "name": "Tejal",
    "age": 22,
    "city": "Nashik"
}

print(student)
```

```python
#Accessing Values
print(student["name"])
```

### Features
Mutable

Ordered (Python 3.7+)

Keys are unique

Values can be duplicated

### Real-life Example

Student records

Employee database

Product information
